## 1. INTRODUCTION

<h4>Problem Statement</h4>

Each baseball season, the MLB hosts an online game called Beat The Streak in which participants select a player each day that they think will get a hit in their MLB game. The participant attempts “to establish a virtual ‘hitting streak’ and keep it going for as long as possible”(Beat the Streak Official Rules). If the player the participant chose gets a hit, their streak will continue, however, if they do not get a hit, the streak will go back down to zero. The goal is to get up to a streak of 57 days in order to beat Joe DiMaggio’s 56-game hitting streak. We want to create a program that will use the data on all players and games in the MLB in order to output a ranked list of the top 5 players that are most likely to get a hit on a given day with their calculated probability of getting a hit. 

<h4>Significance of the Problem</h4>

The winner of this competition wins $5.6 million, and no one has ever successfully completed the 57-day objective, so we want to use predictions based on the provided MLB data in order to try to beat the game. The insights from this problem could end up being a helpful and applicable tool for team managers in the MLB who are trying to choose which players to put in the lineup based on their probability of getting a hit in that game. Some previous work on this topic can be found on the following websites.

```python
import pandas as pd
hitters = pd.read_csv('https://raw.githubusercontent.com/kevschmitt/Beat-the-Streak/main/hittersByGame.csv')
hitters = hitters[['Hitters', 'AB', 'H', 'Game']]
games = pd.read_csv('https://raw.githubusercontent.com/kevschmitt/Beat-the-Streak/main/games.csv')
games = games[['Game', 'Date']]
hitters.head()
```

```python
b_v_p = pd.read_csv('https://raw.githubusercontent.com/kevschmitt/Beat-the-Streak/main/batter_vs_pitcher.csv')
b_v_p
```

```python
def format_dates_bvp(date):
    parts = date.split('/')
    ret = ''
    for p in range(0,2):
        if len(parts[p]) == 1:
            ret = ret + '0' + parts[p] + '/'
        else:
            ret = ret + parts[p] + '/'
    ret = ret + parts[2]
    return(ret)
```

```python
b_v_p['Date'] = b_v_p['Date'].map(format_dates_bvp)
```

```python
games_stats = pd.merge(hitters, games, on='Game', how='left')
```

```python
def format_date(date):
    temp = date.split('T')[0]
    parts = temp.split('-')
    return parts[1] + '/' + parts[2] + '/' + parts[0]
```

```python
def format_name(name):
    parts = name.split(' ')
    fi = parts[0][:1]
    return fi + '. ' + parts[1]
```

```python
games_stats['Date'] = games_stats['Date'].map(format_date)
games_stats.head()
```

```python
b_v_p['Name'] = b_v_p['Name'].map(format_name)
b_v_p = b_v_p.rename(columns = {'Name': 'Hitters'})
b_v_p.head()
```

```python
final = pd.merge(b_v_p, games_stats, on=['Hitters', 'Date'], how='left')
```

```python
final = final.dropna()
final=final.reset_index()
```

```python
final = final[['Hitters', 'Pitcher', 'Team', 'Opp', 'Home?', 'Date', 'AB_x', 'H_x', 'XBH', 'HR', 'RBI', 'BB', 'AVG',
               'OBP', 'AB_y', 'H_y']]
final = final.rename(columns = {'Hitters': 'Hitter', 'AB_x': 'ABVP', 'H_x': 'HVP', 'AB_y': 'AB_actual', 'H_y': 'H_actual'})
```

```python
final['Pitcher'] = final['Pitcher'].map(format_name)
```

```python
def reformat_home(home):
    if home == 'Yes':
        return True
    return False
```

```python
final['Home?'] = final['Home?'].map(reformat_home)
```

```python
for i in ['ABVP', 'HVP', 'XBH', 'HR', 'RBI', 'BB', 'AVG', 'OBP', 'AB_actual', 'H_actual']:    
    final[i] = final[i].astype(float)
final.head()
```
